```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
sentences = [
    "Sally sells seashells by the seashore, swiftly securing slippery shells for seaside strollers.",
    "Walter waited while Wendy weaved whimsical waltzes with Walter's wobbly walrus.",
    "Chester chased a cheetah, challenging his chosen champion to a cheery cheerleading championship."
]
embeddings = model.encode(sentences)

#Print the embeddings
for sentence, embedding in zip(sentences,embeddings):
  print("###############################################################\n###############################################################")
  print("")

  print("Sentence:", sentence)
  print("Embedding:", embedding)
  print("")
```

- Pick a sentence with one or more words with multiple meanings (depending on context), like “flies” (could mean the plural - the insect -  or could be the verb “to fly”), and compute the contextual BERT embeddings for all the the words in the sentence.

- Pick another sentence, with one or more of those ambiguous words but with another meaning, and generate the contextual BERT embeddings for the words in that sentence.  Compare and contrast contextual embedding vectors of words which have changed meaning and those whose meaning has stayed the same.


NB: this code does the embedding for whole sentence instead of individual words.
The next block of code does it rightly.

```python
from scipy.spatial.distance import cosine
from sentence_transformers import SentenceTransformer

# Initialize the model with bert-base-uncased
model = SentenceTransformer('bert-base-uncased')

# Define sentences with words that have multiple meanings
sentence1 = "He cannot bear the weight."
sentence2 = "There is a bear ahead of us."

# Generate embeddings for the sentences
embeddings1 = model.encode([sentence1])
embeddings2 = model.encode([sentence2])

# Extract embeddings for all words in each sentence
words1 = sentence1.split()
words2 = sentence2.split()

embeddings_word1 = {word: model.encode([word])[0] for word in words1}
embeddings_word2 = {word: model.encode([word])[0] for word in words2}

# Compare and contrast contextual embeddings
ambiguous_word = "bear"

print(f"Embedding for the word '{ambiguous_word}' in the sentence '{sentence1}':")
print(embeddings_word1[ambiguous_word])
print("\n")

print(f"Embedding for the word '{ambiguous_word}' in the sentence '{sentence2}':")
print(embeddings_word2[ambiguous_word])
print("\n")

# Compute and print the cosine similarity between the embeddings of the ambiguous word
cos_similarity = 1 - cosine(embeddings_word1[ambiguous_word], embeddings_word2[ambiguous_word])
print(f"Cosine similarity between '{ambiguous_word}' in the two sentences: {cos_similarity}")
```

Here is the assignment proper

```python
from scipy.spatial.distance import cosine
from transformers import BertTokenizer, BertModel
import torch

# Initialize the BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

def get_embedding(sentence):
    # Tokenize the sentence and convert tokens to token IDs
    inputs = tokenizer(sentence, return_tensors="pt", padding=True, truncation=True)

    # Get the output from the BERT model
    with torch.no_grad():
        outputs = model(**inputs)

    # Extract the embeddings for each token
    embeddings = outputs.last_hidden_state[0].numpy()

    return embeddings

# Define the sentences
sentence1 = "They left to go to the market."
sentence2 = "This is my left hand."

# Get embeddings for tokens in each sentence
embeddings1 = get_embedding(sentence1)
embeddings2 = get_embedding(sentence2)

# Convert the tokens to IDs for both sentences
input_ids1 = tokenizer.encode(sentence1, add_special_tokens=True)
input_ids2 = tokenizer.encode(sentence2, add_special_tokens=True)

# Locate the index of the token "left" in the tokenized sentences
index_left = input_ids1.index(tokenizer.convert_tokens_to_ids("left"))

# Extract embeddings for the token "left" from both sentences using the identified index
embedding_left_1 = embeddings1[index_left]
embedding_left_2 = embeddings2[index_left]

# Compute cosine similarity between the embeddings
cos_similarity = 1 - cosine(embedding_left_1, embedding_left_2)

print(f"Cosine similarity for the token 'like' between the two sentences: {cos_similarity}")
```
